```python
d = dict()
all_words=[""]

for line in text: 
    #print(line)
    words =line.split()
    all_words.extend(words)
    #print(words)
    #print("**********************************************************************************************")
    for word in words: 
        # Check if the word is already in dictionary 
        if word in d: 
            # Increment count of word by 1 
            d[word] = d[word] + 1
        else: 
            # Add the word to dictionary with count 1 
            d[word] = 1
            

print("all word number:")
print(len(all_words))
print("uniq word number:")
print(len(list(d)))


with open('all_word_contain.txt', 'w', encoding="utf-8") as fobj:
    for k, v in d.items():
        fobj.write('{}: {}\n'.format(k, v))


"""
import _pickle as pickle
with open('all_word_container.txt', 'wb') as file:
    file.write(pickle.dumps(d)) 

file = open('all_word_container.txt', 'w')
pickle.dump(d, file)
file.close()


for key in list(d.keys()): 
    #all_word_container.write(%key ":" %d[key])
    print(key, ":", d[key]) 
"""
```
